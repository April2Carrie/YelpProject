```python
import pandas as pd
import numpy as np
import csv
```

```python
review = pd.read_csv("review_chinsese_philadelphia.csv")
business = pd.read_csv("business_chinese_philadelphia.csv")
review
```

```python
business_need = business[['business_id','name','address','review_count','hours','is_open']]
```

```python
merged_data = pd.merge(review, business_need, on=["business_id"], how = 'left')
final_data = merged_data[merged_data['is_open'] ==1]
final_data
```
